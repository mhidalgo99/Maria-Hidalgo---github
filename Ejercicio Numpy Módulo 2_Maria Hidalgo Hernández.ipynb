## Tukey y Outliers

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

Q1_price, Q3_price = np.percentile(price, [25, 75])
IQR_price = Q3_price - Q1_price

Q1_rooms, Q3_rooms = np.percentile(rooms, [25, 75])
IQR_rooms = Q3_rooms - Q1_rooms

Q1_baths, Q3_baths = np.percentile(baths, [25, 75])
IQR_baths = Q3_baths - Q1_baths

Q1_sqft, Q3_sqft = np.percentile(sqft, [25, 75])
IQR_sqft = Q3_sqft - Q1_sqft

lower_price, upper_price = Q1_price - 1.5 * IQR_price, Q3_price + 1.5 * IQR_price
lower_rooms, upper_rooms = Q1_rooms - 1.5 * IQR_rooms, Q3_rooms + 1.5 * IQR_rooms
lower_baths, upper_baths = Q1_baths - 1.5 * IQR_baths, Q3_baths + 1.5 * IQR_baths
lower_sqft, upper_sqft = Q1_sqft - 1.5 * IQR_sqft, Q3_sqft + 1.5 * IQR_sqft

outliers_price = price[(price < lower_price) | (price > upper_price)]
outliers_rooms = rooms[(rooms < lower_rooms) | (rooms > upper_rooms)]
outliers_baths = baths[(baths < lower_baths) | (baths > upper_baths)]
outliers_sqft = sqft[(sqft < lower_sqft) | (sqft > upper_sqft)]

print(f"Outliers detectados:")
print(f"Precio: {len(outliers_price)}")
print(f"Habitaciones: {len(outliers_rooms)}")
print(f"Baños: {len(outliers_baths)}")
print(f"Metros cuadrados: {len(outliers_sqft)}")

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.boxplot(x=price)
plt.title("Outliers en Precio")
plt.axvline(lower_price, color='red', linestyle='--', label="Límite inferior")
plt.axvline(upper_price, color='blue', linestyle='--', label="Límite superior")
plt.legend()

plt.subplot(2, 2, 2)
sns.boxplot(x=rooms)
plt.title("Outliers en Habitaciones")
plt.axvline(lower_rooms, color='red', linestyle='--', label="Límite inferior")
plt.axvline(upper_rooms, color='blue', linestyle='--', label="Límite superior")
plt.legend()

plt.subplot(2, 2, 3)
sns.boxplot(x=baths)
plt.title("Outliers en Baños")
plt.axvline(lower_baths, color='red', linestyle='--', label="Límite inferior")
plt.axvline(upper_baths, color='blue', linestyle='--', label="Límite superior")
plt.legend()

plt.subplot(2, 2, 4)
sns.boxplot(x=sqft)
plt.title("Outliers en Metros cuadrados")
plt.axvline(lower_sqft, color='red', linestyle='--', label="Límite inferior")
plt.axvline(upper_sqft, color='blue', linestyle='--', label="Límite superior")
plt.legend()

plt.tight_layout()
plt.show()
```

## Matriz de Correlación

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore

data = np.genfromtxt('/Users/mariahidalgo/Desktop/Maria Hidalgo - github/Data/idealista_madrid.csv',                      delimiter=',', dtype="str", encoding="utf-8", skip_header=1)

price = data[:, 0].astype(float)
rooms = data[:, 1].astype(float)
baths = data[:, 2].astype(float)
sqft = data[:, 3].astype(float)

valid_mask = (price > 0) & (rooms > 0) & (baths > 0) & (sqft > 0)  # Descarta valores negativos o ceros si son inválidos
price = price[valid_mask]
rooms = rooms[valid_mask]
baths = baths[valid_mask]
sqft = sqft[valid_mask]

data_cleaned = np.array([price, rooms, baths, sqft]).T  # Transponer para organizar por filas

correlation_matrix = np.corrcoef(data_cleaned, rowvar=False)  # rowvar=False para correlación entre columnas

print("Matriz de correlación:")
print(correlation_matrix)

plt.figure(figsize=(8, 6))  # Tamaño del gráfico
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', 
            xticklabels=["Price", "Rooms", "Baths", "Sqft"], 
            yticklabels=["Price", "Rooms", "Baths", "Sqft"])
plt.title("Matriz de correlación")
plt.show()
```

## Asímetria y Curtosis 

```python
def calculate_skew(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    n = len(data)
    skewness = np.sum(((data - mean) / std_dev) ** 3) / n
    return skewness

skew_price = calculate_skew(price)
skew_rooms = calculate_skew(rooms)
skew_baths = calculate_skew(baths)
skew_sqft = calculate_skew(sqft)

print(f"Asimetría:")
print(f"Precio: {skew_price}, Habitaciones: {skew_rooms}, Baños: {skew_baths}, Metros Cuadrados: {skew_sqft}")

def calculate_kurtosis(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    n = len(data)
    kurtosis = np.sum(((data - mean) / std_dev) ** 4) / n - 3
    return kurtosis

kurt_price = calculate_kurtosis(price)
kurt_rooms = calculate_kurtosis(rooms)
kurt_baths = calculate_kurtosis(baths)
kurt_sqft = calculate_kurtosis(sqft)

print(f"Curtosis:")
print(f"Precio: {kurt_price}, Habitaciones: {kurt_rooms}, Baños: {kurt_baths}, Metros Cuadrados: {kurt_sqft}")
```

## Contraste de Hipótesis

### Primer contraste
### H0: las casas en el Barrio de Salamanca tienen un precio igual o menor que en Hortaleza.
### H1: las casas en el Barrio de Salamanca tienen un precio mayor que en Hortaleza.

```python
from scipy.stats import ttest_ind

addresses = data[:, 4]  

barrio_x = "Barrio de Salamanca"
barrio_y = "Hortaleza"  

price = data[:, 0].astype(float)  
price_x = price[addresses == barrio_x]
price_y = price[addresses == barrio_y]

if len(price_x) == 0 or len(price_y) == 0:
    print("Error: Uno o ambos barrios no tienen datos. Verifica los nombres de los barrios.")
else:

    t_stat, p_value = ttest_ind(price_x, price_y, alternative='less')  

    print(f"Contraste de hipótesis: Las casas de {barrio_x} son más baratas que las de {barrio_y}")
    print(f"T-statistic: {t_stat}")
    print(f"P-value: {p_value}")

    if p_value < 0.05:
        print(f"Rechazamos H0: Las casas de {barrio_x} son más baratas que las de {barrio_y}.")
    else:
        print(f"No rechazamos H0: No hay evidencia suficiente para afirmar que {barrio_x} sea más barato.")
```
